```python
import pandas as pd

results = []

paths = {
    "A": r"C:\evidence_A\notification.log",
    "B": r"C:\evidence_B\notification.log",
    "C": r"C:\evidence_C\notification.log",
    "D": r"C:\evidence_D\notification.log",
    "E": r"C:\evidence_E\notification.log",
    "F": r"C:\evidence_F\notification.log"
}

for s, path in paths.items():
    try:
        with open(path, "r", encoding="utf-8", errors="ignore") as f:
            content = f.read().lower()
        
        whatsapp_found = "yes" if "whatsapp" in content else "no"
        message_found = "yes" if "test message" in content else "no"
        
        results.append({
            "Scenario": s,
            "WhatsApp_Notification": whatsapp_found,
            "Message_Content_Leak": message_found
        })
        
    except:
        results.append({
            "Scenario": s,
            "WhatsApp_Notification": "not found",
            "Message_Content_Leak": "not found"
        })

df = pd.DataFrame(results)

print(df)

df.to_csv(r"C:\Users\HP\Analysis\notification_results.csv", index=False)
```
